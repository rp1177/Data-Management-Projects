```python
#import libraries:

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#read the excel file into a dataframe format:

google = pd.read_excel('GooglePlaystore.xlsx')
google
```

#### Preprocessing (28 pts)

1.) Often there are outliers which do not match the overall data type. There is one record in this data where the "Reviews" has value "3.0M" which does not match the rest of the data.

```python
outlier = google[google['Reviews'] == '3.0M']
google = google.drop(outlier.index)
google
```

2.)  Remove rows where any of the columns has the value "Varies with device".


```python
outlier2 = 'Varies with device'
found = google[google.apply(lambda x: outlier2 in x.values, axis=1)]
google = google.drop(found.index)
google
```

3.) The values in the Android version column should be floats. Strip the trailing non-numeric characters from all values (ie. the words " and up"), so the result is a number. If there are multiple decimal places (eg. "x.y.z"), keep only the first two parts (eg "x.y"). For example, the value "4.1 and up" should be changed to "4.1". The value "4.5.6 and up" should be changed to "4.5". The value "5.6.7" should be changed to "5.6".

If there is a range (eg. 5.0 - 8.0), only consider the first number. For example, the value "5.0 - 8.0" should be changed to "5.0". The value "4.0.3 - 7.1.1" should be changed to "4.0".

```python
s = google['Android Ver'].str.replace('[A-Za-z ]+', '', regex=True)
s = s.str.extract('(\d\.\d)')

google['Android Ver'] = s

google['Android Ver'] = google['Android Ver'].astype(float)
```

4.) The "Installs" column must have integer values. For values that have commas, remove the commas. For values that have a '+' at the end, remove the '+'. Keep only those rows that have an integer value after these edits.

```python
installs = google['Installs'].str.replace(',', '')
installs = installs.str.replace('+', '')
installs = installs.astype(int)

google['Installs'] = installs
google
```

5.) For missing rating values, if the number of reviews is less than 100 and installations is less than 50000, remove the row. Else, fill the missing value with the average value (rounded to 2 decimal places) for the Category of that row.

```python
#noticing reviews is an object, convert it to integers
google['Reviews'] = google['Reviews'].astype(int)

#if the number of reviews is less than 100 and installations is less than 50000, remove the row.
missing_val = google[(google['Rating'].isnull()) & (google['Reviews'] < 100) & (google['Installs'] < 50000)]
google = google.drop(missing_val.index)

#else... ratings average for everything else that has NaN ratings:
cat_avg = google.groupby('Category')['Rating']
google['Rating'] = google['Rating'].fillna(round(cat_avg.transform('mean'),2))

google
```

6.) Preprocess the Size column to convert the "M" (millions) and "K" (thousands) values into integers. For instance, 8.7M should be converted to 8700000 and 2.4K should be converted to 2400.

```python
def convert(obj):
    multiplier = 1
    if 'k' in obj:
        multiplier = 1000
        obj = obj.replace('k', '')
    elif 'M' in obj:
        multiplier = 1000000
        obj = obj.replace('M', '')
    return int(float(obj) * multiplier)


s = google['Size'].apply(convert)
google['Size'] = s

google


```

2.) Extract all "Free" apps from the master data frame. Then write a function that, given a numeric column e.g 'Rating'), will create and return a dataframe for the top 3 free applications in each category based on that column. Call the function on each of these columns:
-Rating (gives top 3 most highly rated applications in each category)
-Installs (gives top 3 most installed applications in each category)
-Reviews (gives top 3 most reviewed applications in each category)

```python
#Extract all "Free" apps from the master data frame. 
free_apps = google[google['Type'] == 'Free']
free_apps
    
```

```python
#function that gets the top 3 free applications in a numeric column.
def top_three(df, col):
    
    s = df.groupby('Category')[col].nlargest(3)
    s = s.reset_index() #make it into a dataframe.
    s = s.merge(df[['App']], left_on='level_1', right_index=True)
    s = s.drop('level_1', axis=1)
    
    s = s[['Category', 'App', col]] #rearrange
    
    return s
```

a.) Rating (gives top 3 most highly rated applications in each category)


```python
top_three(free_apps, 'Rating')
```

```python
top_three(free_apps, 'Installs')
```

```python
top_three(free_apps, 'Reviews')
```

3.) Find the average, maximum and minimum price of the paid applications.

```python
#first extract the Paid versions of apps in Type, convert to float:


paid_apps = google[google['Type'] == 'Paid'].copy()
paid_apps['Price'] = paid_apps['Price'].astype(float) 

#finding the average of paid applications:
print("Average price of paid applications:", paid_apps['Price'].mean())

#maximum price of paid applications:
print("Maximum price out of all paid applications:", paid_apps['Price'].max())

#minimum price of paid applications:
print("Minimum price out of all paid applications:", paid_apps['Price'].min())
```

#### Visualization

1.) In the genre column, break the string of genres into a list. For example, ‘Art & Design; Creativity’ should be [‘Art & Design’, Creativity’].
Count the number of applications per genre and display it using a pie chart.
Hint: Read about DataFrame.explode()



```python
genres = google[['Genres', 'App']]
genres['Genres'] = genres['Genres'].str.split(';')
genres = genres.explode('Genres')
#genres

count_by_app = genres.groupby('Genres')['App'].count().sort_values(ascending=False)
count_by_app

plt.figure(figsize=(25, 18))
plt.title('Number of Applications Per Genre', fontsize=14)

plt.pie(count_by_app, labels=count_by_app.index, autopct='%1.1f%%', rotatelabels=True)
plt.show()
```
